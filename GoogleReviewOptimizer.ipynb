```python
import pandas as pd
import string
import re
import nltk
from num2words import num2words
from nltk.util import ngrams
nltk.download('stopwords')
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
nltk.download('wordnet')
nltk.download('omw-1.4')

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

stop_words=set(nltk.corpus.stopwords.words('english'))


```

```python
appname=["de.markusfisch.android.screentime",'com.screentime.rc','com.tryohana.android',
        'com.mcafee.security.safefamily','de.lhoer0.screentimeout','com.kiddoware.kidsplace',
        'master.app.screentime','com.burockgames.timeclocker','com.appbuck3t.screentime','com.agooday.screentime',
        'com.ezlapp.screen.time']
```

```python
count=0
stopwords=nltk.corpus.stopwords.words('english')
pd.options.display.max_colwidth=15
Reviews=[]


#Function to change num to text
def numtext(review):
    words=review.split()
    temp=0
    while(temp<len(words)):
        if words[temp].isnumeric():
            words[temp]=num2words(words[temp])
        temp=temp+1
    review=' '.join(words)
    return review

#Function to remove stopwords
def stpwords(review):
    words=review.split()
    words=[word for word in words if word not in stopwords]
    review=' '.join(words)
    return review


#Funtion to Lemmatize the Reviews
def lemm(review):
    lemmatizer = WordNetLemmatizer()
    words=review.split()
    lemmentized=[lemmatizer.lemmatize(word) for word in words]
    review=' '.join(lemmentized)
    return review



#Loop to preprocess the Reviews
while(count<len(appname)):
    file=pd.read_csv('/Users/musta/4312Project/'+appname[count]+'.csv')

    file["Review"]=file["Review"].apply(lambda row: "".join([char for char in str(row) if char not in string.punctuation]))
    file["Review"]=file["Review"].apply(lambda row: re.sub(r"[^a-zA-Z0-9 ]","",row))
    file["Review"]=file["Review"].apply(numtext)
    file["Review"]=file["Review"].apply(lambda row: ' '.join(row.split()))
    file["Review"]=file["Review"].apply(lambda row: row.lower())
    file["Review"]=file["Review"].apply(stpwords)
    file["Review"]=file["Review"].apply(lemm)
    file.to_csv('/Users/musta/4312Project/'+appname[count]+'.csv',index=False)
    print(file.head(15))
    count=count+1
    
    


```
